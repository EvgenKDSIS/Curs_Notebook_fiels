## y" + y' - 2y = 0 
## y(0) = 1
## y(1) = 2

```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

## Данные


```python
x0 = 0
x1 = 1
x_end = 2
C = 1
N = 100
x = np.linspace(x0, x1, N)
```

## Итоговое уравнение

```python
y = C * np.exp(-2 * x - 3) * (np.exp(3 * x) - np.exp(3))
```

```python
# Данные
h = 00000.1
xArr = []
yArr = []

while x0 <= x1:
    # equation
    y_new = C * np.exp(-2 * x0 - 3) * (np.exp(3 * x0) - np.exp(3))
    xArr.append(x0)
    yArr.append(y_new)
    x0 += h
```

```python
# Данные
x_bgn_krv = x0
y_bgn_krv = C * np.exp(-2 * x_bgn_krv - 3) * (np.exp(3 * x_bgn_krv) - np.exp(3))
x_end_krv = x_end
y_end_krv = C * np.exp(-2 * x_end_krv - 3) * (np.exp(3 * x_end_krv) - np.exp(3))

a2 = 1
a1 = 0
a0 = -1

l = x_end_krv - x_bgn_krv
n = 100
h_kraev = l / n
```

```python
alfa = 1 / m.pow(h_kraev, 2)
beta = 2 / m.pow(h_kraev, 2) + 1 / h_kraev - 2
gama = 1 / m.pow(h_kraev, 2) - 1 / h_kraev
```

```python
i = 1
ArrAi = [0]
ArrBi = [y_bgn_krv]

xKRV = [x_bgn_krv]
yKRV = [y_end_krv]

while(i < n):
    xKRV.append(xKRV[i-1] + h_kraev)
    yKRV.append(0)
    ArrAi.append(-alfa / ((beta + gama * ArrAi[i-1])))
    ArrBi.append((-gama * ArrBi[i-1]) /(beta + gama * ArrAi[i-1]))
    i += 1
    
    
i -= 1
yKRV[i] = y_end_krv
i -= 1

while(i >= 0):
    yKRV[i] = (ArrAi[i]*(yKRV[i+1]) + ArrBi[i])
    i -= 1
```

```python
# печать графиков
plt.title("решение уравнения")
plt.plot(x, y, color="red", linestyle='solid', label='Аналитическое решение')
plt.plot(xArr, yArr, color="yellow", linestyle='dashed')
plt.plot(xKRV, yKRV, color="blue", linestyle='dashed', label='Метод прогонки')
plt.legend(loc='upper left')
plt.grid()
```
