### метод Рунге-Кутты

```python
# функция

def f11(fx,fy):
  return ( ( np.exp(fx + 1) + np.exp(-fx + 1) ) / ( np.exp(2) - 1 ) ) - 2
```

```python
# Данные
h = 0.001
i = 0

xRan = [x0]
yRan = [( ( np.exp(x0 + 1) - np.exp(-x0 + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x0]
```

```python
while xRan[i] <= x_end:

  k1 = f11(xRan[i],yRan[i])
  k2 = f11(xRan[i]+h/2, yRan[i]+k1*h/2)
  k3 = f11(xRan[i]+h/2, yRan[i]+k2*h/2)
  k4 = f11(xRan[i]+h, yRan[i]+k3*h)

  xRan.append(xRan[i] + h)
  yRan.append(yRan[i] +  h*(k1 + 2*k2 + 2*k3 + k4) /6 )

  i += 1
```

### Краевые виды задач

```python
# Данные
x_bgn_krv = x0
y_bgn_krv = ( ( np.exp(x_bgn_krv + 1) - np.exp(-x_bgn_krv + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x_bgn_krv
x_end_krv = x_end
y_end_krv = ( ( np.exp(x_end_krv + 1) - np.exp(-x_end_krv + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x_end_krv

a2 = 1
a1 = 0
a0 = -1

l = x_end_krv - x_bgn_krv
n = 500
h_kraev = l / n

alfa = a2 / (h_kraev**2) - a1 / h_kraev
beta = - (2 * a2) / (h_kraev**2) + a1 / h_kraev + a0
gama = a2 / (h_kraev**2)
```

```python
i = 1
ArrAi = [0]
ArrBi = [y_bgn_krv]

xKRV = [x_bgn_krv]
yKRV = [y_end_krv]

while(i < n):
  xKRV.append(xKRV[i-1] + h_kraev)
  yKRV.append(0)

  ArrAi.append( -gama / ( (alfa * ArrAi[i-1]) + beta) )
  ArrBi.append( ( 2*xKRV[i] - (alfa * ArrBi[i-1])) / ( (alfa * ArrAi[i-1]) + beta) )

  i += 1
i -= 1
yKRV[i] = y_end_krv
i -= 1

while(i >= 0):
  yKRV[i] = (ArrAi[i]*(yKRV[i+1]) + ArrBi[i])

  i -= 1
```

```python
# Данные

x0 = -2
x_end = 2
x = np.linspace(x0, x_end, 100)
```

```python
# equation
y = ( ( np.exp(x + 1) - np.exp(-x + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x
```

### метод Эйлера

```python
# Данные
dx = 0.01

xArr = []
yArr = []

x_new = x0
```

```python
while x_new <= x_end:

  # equation
  y_new = ( ( np.exp(x_new + 1) - np.exp(-x_new + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x_new


  xArr.append(x_new)
  yArr.append(y_new)

  x_new += dx
```

### метод Рунге-Кутты

```python
# функция

def f11(fx,fy):
  return ( ( у.exp(fx + 1) + np.exp(-fx + 1) ) / ( np.exp(2) - 1 ) ) - 2
```

```python
# Данные
h = 0.001
i = 0

xRan = [x0]
yRan = [( ( np.exp(x0 + 1) - np.exp(-x0 + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x0]
```

```python
while xRan[i] <= x_end:

  k1 = f11(xRan[i],yRan[i])
  k2 = f11(xRan[i]+h/2, yRan[i]+k1*h/2)
  k3 = f11(xRan[i]+h/2, yRan[i]+k2*h/2)
  k4 = f11(xRan[i]+h, yRan[i]+k3*h)

  xRan.append(xRan[i] + h)
  yRan.append(yRan[i] +  h*(k1 + 2*k2 + 2*k3 + k4) /6 )

  i += 1
```

### Краевые виды задач

```python
# Данные
x_bgn_krv = x0
y_bgn_krv = ( ( np.exp(x_bgn_krv + 1) - np.exp(-x_bgn_krv + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x_bgn_krv
x_end_krv = x_end
y_end_krv = ( ( np.exp(x_end_krv + 1) - np.exp(-x_end_krv + 1) ) / ( np.exp(2) - 1 ) ) - 2 * x_end_krv

a2 = 1
a1 = 0
a0 = -1

l = x_end_krv - x_bgn_krv
n = 100
h_kraev = l / n

alfa = a2 / (h_kraev**2) - a1 / h_kraev
beta = - (2 * a2) / (h_kraev**2) + a1 / h_kraev + a0
gama = a2 / (h_kraev**2)
```

```python
i = 1
ArrAi = [0]
ArrBi = [y_bgn_krv]

xKRV = [x_bgn_krv]
yKRV = [y_end_krv]

while(i < n):
  xKRV.append(xKRV[i-1] + h_kraev)
  yKRV.append(0)

  ArrAi.append( -gama / ( (alfa * ArrAi[i-1]) + beta) )
  ArrBi.append( ( 2*xKRV[i] - (alfa * ArrBi[i-1])) / ( (alfa * ArrAi[i-1]) + beta) )

  i += 1
i -= 1
yKRV[i] = y_end_krv
i -= 1

while(i >= 0):
  yKRV[i] = (ArrAi[i]*(yKRV[i+1]) + ArrBi[i])

  i -= 1
```
