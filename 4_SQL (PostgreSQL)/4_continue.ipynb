```python
from faker import Faker
import random

fake = Faker()

# Generate and print the CREATE TABLE statement
create_table_sql = """
CREATE TABLE fake_data_2 (
    id serial PRIMARY KEY,
    name varchar(50),
    age integer,
    gender varchar(10),
    occupation varchar(50),
    salary float
);
"""

print(create_table_sql)

# Generate and print the INSERT INTO statements
insert_statements = []
genders = ['Male', 'Female']

occupations = [
    'Software Developer',
    'Data Analyst',
    'Teacher',
    'Nurse',
    'Engineer',
    'Designer',
    'Sales Manager',
    'Writer'
]

for _ in range(1000):
    name = fake.name()
    age = random.randint(18, 90)
    gender = random.choice(genders)
    occupation = random.choice(occupations)
    salary = round(random.uniform(30000, 100000), 2)

    insert_statement = f"INSERT INTO fake_data_2 (name, age, gender, occupation, salary) VALUES ('{name}', {age}, '{gender}', '{occupation}', {salary});"
    insert_statements.append(insert_statement)

for statement in insert_statements:
    print(statement)
```
